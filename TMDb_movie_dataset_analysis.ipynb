## Looking at the dataset we can find that:<a href="#Looking-at-the-dataset-we-can-find-that:"
class="anchor-link">¶</a>

-   The columns ' *budget* ',' *revenue* ',' *budget_adj* ','
    *revenue_adj* ' have not been mentioned with any currency but It was
    clearly given in the **Instruction and notes** section that the
    *\_adj* columns of associated movie are in terms of 2010 dollars,
    accounting for inflation over time.So,the currency is in dollars.
-   The vote count for all the movies are not the same,they differ each
    and every movie.So we cannot assume the movie with the highest votes
    has the greatest success rate or vice versa.
-   There are many columns that are not needed for the analysis and must
    be removed.They are ' **imdb_id** ',' **homepage** ','**tagline**
    ',' **keywords** ',' **overview** ',' **vote_count** ','
    **budget_adj** ',' **revenue_adj** ' and ' **vote_average** '.
-   Repeated or duplicate rows if there are any must be removed.
-   Release date must be modified into date format.
-   zeros in some of the columns must be taken care of.

        least=tmdb[column].idxmin()
        least_movie=pd.DataFrame(tmdb.loc[least])

        return lowest_highest
